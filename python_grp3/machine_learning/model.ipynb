# Start Modeling

```python
seed = 42

# create a random variable to check feature importance
bike_data['random'] = np.random.rand(len(bike_data))

X = bike_data.drop(columns=['total_bike_ct','casual_user_ct','registered_user_ct'])

y = bike_data['total_bike_ct']

X.shape
```

```python
labelenc_val_cat = ['year', 'is_holiday', 'is_workingday', 'hourly_rental_deviation_label', 'weather_condition']
print("Label Encoding: ", labelenc_val_cat)
ohe_val_cat = ['season', 'day_part']
print("One Hot Encoding: ", ohe_val_cat)
#take everything apart from target variable + casual and registered user count
num_vars = bike_data.select_dtypes(include=['float', 'int']).drop(['total_bike_ct', 'casual_user_ct', 'registered_user_ct'], axis=1).columns
print("Numerical Variables: ", num_vars)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import RobustScaler

# Create a column transformer

preprocessor = ColumnTransformer(
    transformers=[
        ("cat_labelenc", OrdinalEncoder() , labelenc_val_cat),
        ("cat_ohe",OneHotEncoder(), ohe_val_cat),
        ("num", RobustScaler(), num_vars)
    ],
)

X_prep = preprocessor.fit_transform(X)


print(X.shape)
print(X_prep.shape)
```

```python
from sklearn.model_selection import train_test_split

evalution_window = 7*24 # 7 days * 24 hours = 1week

X_train, X_test, y_train, y_test = train_test_split(X_prep[evalution_window:], 
                                                    y[evalution_window:], 
                                                    test_size=0.3, 
                                                    random_state=seed, shuffle=False)
```

```python
print(X_train.shape)
```

## Baseline Model: Lasso Regression

```python
def mape(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    if np.any(y_true==0):
        print("Warning: Zero values found in y_true, MAPE is undefined. Setting it as NaN")
        return np.nan
    else:
        return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error


# Create a linear regression model with K-fold cross validation
linear_baseline_model = LinearRegression(fit_intercept=True)
linear_baseline_model.fit(X_train, y_train)

basline_preds = linear_baseline_model.predict(X_test)

#check overfitting
y_train_baseline_preds = linear_baseline_model.predict(X_train)

# Create a dataframe to store the scores (MAE, MSE, RMSE, R2) for train, test.
baseline_lr_scores_df = pd.DataFrame(
    {
        "Train": [mean_absolute_error(y_train, y_train_baseline_preds), 
                  mean_squared_error(y_train, y_train_baseline_preds), 
                  np.sqrt(mean_squared_error(y_train, y_train_baseline_preds)), 
                  r2_score(y_train, y_train_baseline_preds), 
                  mape(y_train, y_train_baseline_preds)],
        "Test": [mean_absolute_error(y_test, basline_preds), 
                 mean_squared_error(y_test, basline_preds), 
                 np.sqrt(mean_squared_error(y_test, basline_preds)), 
                 r2_score(y_test, basline_preds),
                 mape(y_test, basline_preds)],
        
        
    },
    index=["MAE", "MSE", "RMSE", "R2","MAPE"]
)

baseline_lr_scores_df
```

```python
column_names = preprocessor.get_feature_names_out()

# Assuming baseline_lr is your fitted model
feature_importance = pd.DataFrame(linear_baseline_model.coef_, index=column_names, columns=['Coefficient']).abs()
feature_importance.sort_values(by='Coefficient', ascending=False, inplace=True)
print(feature_importance)
```

```python
# Plot a scatterplot of the residuals of my lasso model
residuals = y_test - basline_preds
fig = px.scatter(x=y_test, y=residuals, labels={'x':'y_test', 'y':'Residuals'}, title='Residuals vs y_test')
fig.add_shape(type="line", x0=y_test.min(), y0=0, x1=y_test.max(), y1=0, line=dict(color="red", width=3))
fig.show()
```

__Comments__

By doing the baseline, we can see that the model has good generalized (good R2 in the test), however by plotting a scatter plot, we can see that the model's residuals are highly correlated, suggesting that we didn't catch all the linear relationships. I here we can note that a time series forecasting model could be better suited such as an SARIMA or Linear regression with time series residuals, however as this is a machine learning exercice, we will go for a tree base model, which we think can better capture some linear and non-linear relations in our case

## Finding the best Model

__Random Forest Model__

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import TimeSeriesSplit
from sklearn.model_selection import cross_val_score

# Create a time series split
tscv = TimeSeriesSplit(n_splits=3)

# Create a random forest model baseline without hyperparameters
rf_model_base = RandomForestRegressor(random_state=seed)

# Perform cross-validation
cross_val_scores = cross_val_score(rf_model_base, X_train, y_train, cv=tscv, scoring='neg_mean_absolute_error')

print('Cross-validation MAE scores: ', -cross_val_scores)
print('Average cross-validation MAE: ', np.mean(-cross_val_scores))

# Fit the model on the training data
rf_model_base.fit(X_train, y_train)

rf_baseline_preds = rf_model_base.predict(X_test)

rf_baseline_preds_y_train = rf_model_base.predict(X_train)

baseline_rf_scores_df = pd.DataFrame(
    {
        "Train": [
            mean_absolute_error(y_train, rf_baseline_preds_y_train), 
            mean_squared_error(y_train, rf_baseline_preds_y_train), 
            np.sqrt(mean_squared_error(y_train, rf_baseline_preds_y_train)), 
            r2_score(y_train, rf_baseline_preds_y_train),
            mape(y_train, rf_baseline_preds_y_train)
        ],
        "Test": [
            mean_absolute_error(y_test, rf_baseline_preds), 
            mean_squared_error(y_test, rf_baseline_preds), 
            np.sqrt(mean_squared_error(y_test, rf_baseline_preds)), 
            r2_score(y_test, rf_baseline_preds),
            mape(y_test, rf_baseline_preds)
        ],
    },
    index=["MAE", "MSE", "RMSE", "R2", "MAPE"]
)

print(baseline_rf_scores_df)
```

```python
# Get feature importances
feature_importance = pd.DataFrame(rf_model_base.feature_importances_, index=column_names, columns=['Importance']).abs()
feature_importance.sort_values(by='Importance', ascending=False, inplace=True)
feature_importance
```

__Comments__

Here we can see that more than half of our feature doesn't make any difference in the model and are just adding noise! We are going to remove them!

```python
# Remove all the features below the random feature importance
feature_importance_filtered = feature_importance[feature_importance['Importance'] > feature_importance.loc['num__random', 'Importance']]
feature_importance_filtered
```

__Retrain the model with less features__

```python
threshold = 0.003246  # Example threshold
filtered_feature_names = feature_importance[feature_importance['Importance'] > threshold].index.tolist()

# Map filtered feature names to their corresponding indices in X_prep
filtered_feature_indices = [list(column_names).index(name) for name in filtered_feature_names if name in column_names]

# Filter X_prep to keep only the important features
X_prep_filtered = X_prep[:, filtered_feature_indices]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_prep_filtered[evalution_window:], 
                                                    y[evalution_window:], 
                                                    test_size=0.3, 
                                                    random_state=seed, shuffle=False)

# Create a time series split
tscv = TimeSeriesSplit(n_splits=3)

# Create a random forest model baseline without hyperparameters
rf_model_base = RandomForestRegressor(random_state=seed)

# Perform cross-validation
cross_val_scores = cross_val_score(rf_model_base, X_train, y_train, cv=tscv, scoring='neg_mean_absolute_error')

print('Cross-validation MAE scores: ', -cross_val_scores)
print('Average cross-validation MAE: ', np.mean(-cross_val_scores))

# Fit the model on the training data
rf_model_base.fit(X_train, y_train)

rf_baseline_preds = rf_model_base.predict(X_test)

rf_baseline_preds_y_train = rf_model_base.predict(X_train)

baseline_rf_scores_df = pd.DataFrame(
    {
        "Train": [
            mean_absolute_error(y_train, rf_baseline_preds_y_train), 
            mean_squared_error(y_train, rf_baseline_preds_y_train), 
            np.sqrt(mean_squared_error(y_train, rf_baseline_preds_y_train)), 
            r2_score(y_train, rf_baseline_preds_y_train),
            mape(y_train, rf_baseline_preds_y_train)
        ],
        "Test": [
            mean_absolute_error(y_test, rf_baseline_preds), 
            mean_squared_error(y_test, rf_baseline_preds), 
            np.sqrt(mean_squared_error(y_test, rf_baseline_preds)), 
            r2_score(y_test, rf_baseline_preds),
            mape(y_test, rf_baseline_preds)
        ],
    },
    index=["MAE", "MSE", "RMSE", "R2", "MAPE"]
)

print(baseline_rf_scores_df)
```

__Using Grid Search__

```python
from sklearn.model_selection import RandomizedSearchCV, TimeSeriesSplit

tsc = TimeSeriesSplit(n_splits=3)

param_grid = {
    'n_estimators': [100,500,1000],
    'max_depth': [8,9,10],
    'min_samples_leaf': [8,9,10],
    'ccp_alpha': [0.1,0.01,0.001],
}

rs_rf_model = RandomizedSearchCV(RandomForestRegressor(random_state=seed), param_distributions=param_grid, cv=tsc, verbose=2, scoring= 'neg_mean_absolute_error')

rs_rf_model.fit(X_train, y_train)

#get best params
print(rs_rf_model.best_params_)
```

```python
#Get average score of the cross validation search
print(-rs_rf_model.best_score_)
```

```python
rf_best = RandomForestRegressor(**rs_rf_model.best_params_, random_state=seed)

rf_best.fit(X_train, y_train)

rf_preds_test = rf_best.predict(X_test)
rf_preds_train = rf_best.predict(X_train)

rs_rf_scores_df = pd.DataFrame(
    {
        "Train": [mean_absolute_error(y_train, rf_preds_train), 
                  mean_squared_error(y_train, rf_preds_train), 
                  np.sqrt(mean_squared_error(y_train, rf_preds_train)), 
                  r2_score(y_train, rf_preds_train), 
                  mape(y_train, rf_preds_train)],
        "Test": [mean_absolute_error(y_test, rf_preds_test), 
                 mean_squared_error(y_test, rf_preds_test), 
                 np.sqrt(mean_squared_error(y_test, rf_preds_test)), 
                 r2_score(y_test, rf_preds_test),
                 mape(y_test, rf_preds_test)],
        
        
    },
    index=["MAE", "MSE", "RMSE", "R2","MAPE"]
)

rs_rf_scores_df
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score

GaussianNB_model = GaussianNB()

cross_val_scores = cross_val_score(rf_model_base, X_train, y_train, cv=tscv, scoring='neg_mean_absolute_error', verbose=2)

print('Cross-validation MAE scores: ', -cross_val_scores)
print('Average cross-validation MAE: ', np.mean(-cross_val_scores))

GaussianNB_model.fit(X_train, y_train)

GNB_preds_test = GaussianNB_model.predict(X_test)

GNB_preds_y_train = GaussianNB_model.predict(X_train)

GaussianNB_scores_df = pd.DataFrame( 
    {
        "Train": [mean_absolute_error(y_train, GNB_preds_y_train), 
                  mean_squared_error(y_train, GNB_preds_y_train), 
                  np.sqrt(mean_squared_error(y_train, GNB_preds_y_train)), 
                  r2_score(y_train, GNB_preds_y_train), 
                  mape(y_train, GNB_preds_y_train)],
        "Test": [mean_absolute_error(y_test, GNB_preds_test), 
                 mean_squared_error(y_test, GNB_preds_test), 
                 np.sqrt(mean_squared_error(y_test, GNB_preds_test)), 
                 r2_score(y_test, GNB_preds_test),
                 mape(y_test, GNB_preds_test)],
        
        
    },
    index=["MAE", "MSE", "RMSE", "R2","MAPE"]
)

GaussianNB_scores_df
```

__XGBoost__

```python
from xgboost import XGBRegressor

param_grid_xgb = {
    'max_depth': [3,4,5],
    'learning_rate': [0.1,0.01,0.001],
    'n_estimators': [100,500,1000],
}

rs_xgb_model = RandomizedSearchCV(XGBRegressor(random_state=seed), 
                                  param_distributions=param_grid_xgb, 
                                  cv=tsc, 
                                  verbose=2, 
                                  scoring= 'neg_mean_absolute_error')

rs_xgb_model.fit(X_train, y_train)

#get best params
print(rs_xgb_model.best_params_)
```

```python
xgb_best = XGBRegressor(**rs_xgb_model.best_params_, random_state=seed)

xgb_best.fit(X_train, y_train)

xgb_preds_test = xgb_best.predict(X_test)
xgb_preds_train = xgb_best.predict(X_train)

xgb_scores_df = pd.DataFrame(
    {
        "Train": [mean_absolute_error(y_train, xgb_preds_train), 
                  mean_squared_error(y_train, xgb_preds_train), 
                  np.sqrt(mean_squared_error(y_train, xgb_preds_train)), 
                  r2_score(y_train, xgb_preds_train), 
                  mape(y_train, xgb_preds_train)],
        "Test": [mean_absolute_error(y_test, xgb_preds_test), 
                 mean_squared_error(y_test, xgb_preds_test), 
                 np.sqrt(mean_squared_error(y_test, xgb_preds_test)), 
                 r2_score(y_test, xgb_preds_test),
                 mape(y_test, xgb_preds_test)],
        
        
    },
    index=["MAE", "MSE", "RMSE", "R2","MAPE"]
)

xgb_scores_df
```

```python
# Create a dataframe to store the scores (MAE, MSE, RMSE, R2) for train, test of all the models
all_models_scores_df = pd.concat([baseline_lr_scores_df, baseline_rf_scores_df, rs_rf_scores_df, GaussianNB_scores_df, xgb_scores_df], axis=1)
all_models_scores_df.columns = ['BS_LR_Train', 'BS_LR_Test', 
                                'BS_RF_Train', 'BS_RF_Test', 
                                'RS_RF_Train', 'RS_RF_Test', 
                                'GNB_Train', 'GNB_Test', 
                                'RS_XGB_Train', 'RS_XGB_Test']

all_models_scores_df
```
