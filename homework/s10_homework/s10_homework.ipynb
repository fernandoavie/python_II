### Step 1: Load data

```python
import pandas as pd
data = pd.read_csv("data/adult.csv")

data.head()
```

### Step 2: Data audit

```python
data.dtypes
```

### As we see the are a lot of categorical variables, so we need to scale them

```python
from sklearn.preprocessing import LabelEncoder

cat_columns = [col for col in data.select_dtypes(include=['object']).columns]

for cat_col in cat_columns:
    print(cat_col, f"{len(data[cat_col].unique())} categories")
    data[cat_col] = LabelEncoder().fit_transform(data[cat_col])
    
data
```

```python
data_corr = data.corr()
data_corr["income"].sort_values(ascending=False)
```

```python
# Converting the Categorical variables
data["age_group"] = pd.cut(data["age"], bins=3).astype('object')
dumi_df = pd.get_dummies(data['educational-num'], prefix='educational-num', drop_first=True)
dumi_bin = dumi_df.map(lambda x: 1 if x == 1 else 0)
data = data.drop("educational-num", axis=1)
data = data.drop("age", axis = 1)
data = pd.concat([data, dumi_bin], axis=1)
data.sample(5)
```

```python
data["income"].value_counts() #Target unbalanced
```

```python
X = pd.DataFrame(data.drop(columns="income"))
X = pd.get_dummies(X, drop_first=True)
X
```

```python
y = pd.Series(data["income"])
y
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import GridSearchCV
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=0,
    shuffle = True,
    stratify = y
)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#Escalar las variables de object a numerico
from sklearn.preprocessing import StandardScaler
scl = StandardScaler()

#NOTA 1: Instead of `fit` and the `transform`, you can use the `fit_transform` method to do both thing sequentially
X_train_scaled = scl.fit_transform(X_train)
X_test_scaled = scl.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score # because the data is imbalanced

rf = RandomForestClassifier()

rf.fit(X_train_scaled, y_train)

baseline_pred_rf = rf.predict(X_test_scaled)

f1_rf = f1_score(baseline_pred_rf, y_test)

f1_rf
```

```python
from sklearn.model_selection import GridSearchCV

grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

gs_rf = GridSearchCV(
    estimator=rf,
    param_grid=grid_rf,
    scoring='f1',
    cv=3,
    verbose=2
)

gs_rf.fit(X_train_scaled, y_train)
```

```python
gs_rf.best_params_
```

```python
gs_rf.best_score_
```
