Add a new column called Close_Std with the average standard deviation of the previous 4 weeks (current week included)## Exercise 01

Load the GOOG stock price history with `yfinance` library since 01-Jan-2015 

 1. Create a new DataFrame with weekly averaged values of `Close`
 1. Add a new column called `Close_MA` with the weekly moving average of the previous 4 weeks (current week included)
 1. Add a new column called `Close_Std` with the average standard deviation of the previous 4 weeks (current week included) 
 1. Add a new column called `Target` with the shifted `Close` price 1 week. That is, for the week N, the `Target` column should contain the average price corresponding to the N+1 week.
 1. Add a new column called `Target_Bin` with the following binary logic:
    - 0 if the next week's price was lower o equal
    - 1 if the next week's price was higher

```python
import yfinance as yf

goog = yf.download('GOOG', start='2015-01-01', end='2024-01-18')
goog.head()
```

```python
goog = goog[['Close']].resample('W').mean()
```

```python
goog.head()
```

```python
type(goog)
```

```python
#2. Add a new column called Close_MA with the weekly moving average of the previous 4 weeks (current week included)
goog['Close_MA'] = goog['Close'].rolling(window=4, min_periods=1).mean()

goog.head(10)
```

```python
# Add a new column called Close_Std with the average standard deviation of the previous 4 weeks (current week included)

goog['Close_Std'] = goog['Close'].rolling(window=4, min_periods=1).std()

goog.head(10)
```

```python
#Add a new column called Target with the shifted Close price 1 week. That is, for the week N, 
#the Target column should contain the average price corresponding to the N+1 week.
goog['Target'] = goog['Close'].shift(-1)

goog.head()
```

```python
#Add a new column called Target_Bin with the following binary logic:
# 0 if the next week's price was lower o equal
# 1 if the next week's price was higher

goog['Target_Bin'] = (goog['Target'] > goog['Close']).astype(int)

goog.head()
```
