````python
import pandas as pd
data = pd.read_csv("data/adult.csv.zip")

data["age_group"] = pd.cut(data["age"], bins=3).astype('object')
dumi_df = pd.get_dummies(data['educational-num'], prefix='educational-num', drop_first=True)
data = data.drop("educational-num", axis=1)
data = data.drop("age", axis = 1)
data = pd.concat([data, dumi_df], axis=1)
data.sample(5)
````

````python
data.isna().sum()
````

````python
#Mirar los tipos de datos
data.dtypes
````

````python
from sklearn.preprocessing import LabelEncoder

cat_columns = [col for col in data.select_dtypes(include=['object']).columns]

for cat_col in cat_columns:
    print(cat_col, f"{len(data[cat_col].unique())} categories")
    data[cat_col] = LabelEncoder().fit_transform(data[cat_col])
    
data
````

````python
data_corr = data.corr()
data_corr["income"].sort_values(ascending=False)
````

````python
X = pd.DataFrame(data.drop(columns="income"))
X = pd.get_dummies(X, drop_first=True)
X["age_group"]
````

````python
y = pd.Series(data["income"])
y
````

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
````

````python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.3, 
    random_state=0
)

print(f"size X_train{X_train.shape}")
print(f"size X_test{X_test.shape}")
print(f"size y_train{y_train.shape}")
print(f"size y_test{y_test.shape}")
````

````python
#Escalar las variables de object a numerico
from sklearn.preprocessing import RobustScaler
scl = RobustScaler()

#NOTA 1: Instead of `fit` and the `transform`, you can use the `fit_transform` method to do both thing sequentially
X_train_sc = scl.fit_transform(X_train)
X_test = scl.transform(X_test)

# the result is returned as a numpy array, so let's convert it again to a Padnas Df
X_train_sc = pd.DataFrame(X_train_sc, columns=X_train.columns)
````

````python
X_train_sc.describe()
````

````python
model = LogisticRegression(max_iter = 200) ## we need to understand what is this
model.fit(X_train_sc, y_train)
````

````python
# Train and evalute on training set
from sklearn.model_selection import cross_validate
results = cross_validate(model, X_train_sc, y_train, cv=5, scoring=["precision", "recall", "f1"], return_train_score=True)
results
````

````python
results = pd.DataFrame(results)
results
````
